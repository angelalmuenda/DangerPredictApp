```python
crime_weather = pd.merge(new_crime_data, weather_data, how='left', left_on=['weather_time','city','date'], right_on=['Time','Region','Date'])
```

```python
sun_crime = pd.merge(crime_weather, sun_rise_set_data, how='left', left_on=['date'], right_on=['Date'])
```

```python
import datetime
def get_dark_light(x,y,z):
    my_time = datetime.datetime.strptime(x, '%H:%M:%S')
    dawn_time = datetime.datetime.strptime(y, '%H:%M')
    dusk_time = datetime.datetime.strptime(z, '%H:%M')
    if my_time.time() < dawn_time.time() or my_time.time() > dusk_time.time():
        return "night"
    else:
        return "day"
```

```python
sun_crime['dark_light'] = sun_crime.apply(lambda row: get_dark_light(row['time'],row['Sunrise_mil'],row['Sunset_mil']),axis=1)
```
