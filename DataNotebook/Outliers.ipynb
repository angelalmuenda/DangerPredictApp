```python
import pandas as pd

crime_places = pd.read_csv('Normalized.csv', encoding='utf-8')
crime_places.drop(['Unnamed: 0'], axis=1, inplace=True)
crime_places
```

```python
for each in crime_places.columns:
    print("'{}',".format(each))
```

```python
crime_places[crime_places.columns[6:]]
crime = crime_places[crime_places.columns[6:]].copy()
```

```python
crime.columns
```

```python
places = [
'Bank',
'Bar',
'Beach',
'Bridge',
'Cemetery',
'Church',
'Clinic',
'Commercial Building',
'Community Center',
'Convenience Store',
'Fire Station',
'Gas Station',
'Government Office',
'Greenfield',
'Highway',
'Highway Pedestrian',
'Highway Primary',
'Highway Residential',
'Highway Secondary',
'Highway Tertiary',
'Hospital',
'Hotel',
'Industrial Building',
'Mall',
'Marketplace',
'Park',
'Parking Area',
'Pawnshop',
'Pharmacy',
'Police Station',
'Post Office',
'Private Office',
'Recreational Area',
'Residential Building',
'Restaurant',
'Road',
'School',
'Sports Center',
'Tourist Attraction',
'Transport Terminal',
]
```

```python
from numpy import mean
from numpy import std

outlier_indexes = []
outlier_columns = []
for col in crime.columns:
    print('---- {} ----'.format(col))
    print('mean=%.3f stdv=%.3f' % (mean(crime[col]), std(crime[col])))
    data_mean, data_std = mean(crime[col]), std(crime[col])
    # identify outliers
    cut_off = data_std * 3
    print('cut_off=%.3f' % (cut_off))
    lower, upper = data_mean - cut_off, data_mean + cut_off
    print('lower=%.3f upper=%.3f' % (lower, upper))
    outliers = [x for x in crime[col] if x < lower or x > upper]
    print('number of outliers=%.3f' % (len(outliers)))
    if col in places and len(outliers) > 0:
        outlier_columns.append(col)
#     for each in outliers:
#         print(each)
    x=0
    while x < len(crime[col]):
        if crime.loc[x,col] < lower or crime.loc[x,col] > upper:
            outlier_indexes.append(x)
            print("{} --- {}".format(x,crime.loc[x,col]))
        x+=1
```
